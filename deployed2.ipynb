```python
medication_names = []
Indications = []
Wit_Indications = {}
for disease in diseases :
  Indications_Infos = data[disease]
  s_Indications = ''
  for Indications_Info in Indications_Infos :
    medication_names.append(Indications_Info['DrugName'])
    Indications.append(Indications_Info['Indications']+'--'+disease+'--')
    #Indications.append(Indications_Info['Indications'])
    s_Indications += Indications_Info['Indications']
  Wit_Indications[disease]  = s_Indications
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(Indications)
```

```python
def search_medicine(x):
    # تحويل النص المدخل من المستخدم إلى مصفوفة TF-IDF
    #xx = [Wit_Indications[x]]
    xx = [x]
    user_tfidf = vectorizer.transform(xx)

    # حساب قيمة التشابه الكوسيني بين النص المدخل من المستخدم وكل الأعراض
    similarity_scores = cosine_similarity(user_tfidf, tfidf_matrix)

    # الحصول على الفهرس الأعلى للأدوية المتوافقة
    top_indices = np.argsort(similarity_scores, axis=1)[:, -5:]

    # الحصول على أسماء الأدوية المتوافقة
    matching_medications = [medication_names[index] for index in top_indices[0]]
    #matching_medications_set = set(matching_medications)

    return matching_medications
```

```python
matching_medications = search_medicine('disease')
matching_medications
```
